```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import ttest_ind

def adj_light(color, l=1, s=1):
    import matplotlib.colors as mc
    import colorsys
    try:
        c = mc.cnames[color]
    except:
        c = color
    c = colorsys.rgb_to_hls(*mc.to_rgb(c))
    return colorsys.hls_to_rgb(c[0], max(0, min(1, l * c[1])), 
                 max(0, min(1, s * c[2])))

acolor = ["#00CC00", "#D8A03D", "#EC008C",
          'cyan', "#B741DC", "#808285",
          'blue', 'black', 'green', 'red',
          'orange', 'brown', 'pink', 'purple']

def printViolin(df, params=None):
    params = params or {}
    sns.set()
    sns.set_style("white")
    sns.set_style({'xtick.color': '.5', 'ytick.color': '.5', 'axes.labelcolor': '.5'})
    sns.set_context("notebook")

    if 'dpi' in params:
        dpi = params['dpi']
    else:
        dpi = 1200

    w, h = (1.5 * len(df['Staining Protocol'].unique()), 5)
    if 'w' in params:
        w = params['w']
    if 'h' in params:
        h = params['h']

    color_sch1 = acolor
    if 'acolor' in params:
        color_sch1 = params['acolor']

    ax = None
    if 'ax' in params:
        ax = params['ax']

    if ax is None:
        fig, ax = plt.subplots(figsize=(w, h), dpi=dpi)

    

    # Set the color palette
    sns.set_palette([adj_light(c, 1.5, 1) for c in color_sch1])
    larger_width = w * 1.5  
    fig, ax = plt.subplots(figsize=(larger_width, h), dpi=dpi)

    width = 0.95
    height = 1
    if 'width' in params:
        width = params['width']

    if 'vert' in params and params['vert'] == 1:
        ax = sns.violinplot(x='Staining Protocol', y='Brown Intensity(25% top - 25% bottom)', inner='quartile',
                            linewidth=0.2, width=width, ax=ax, data=df,
                            order=['Boiling-2min', 'Boiling-5min', 'Boiling-15min', 'Pressure-3min'], orient='v')
        ax = sns.swarmplot(x='Staining Protocol', y='Brown Intensity(25% top - 25% bottom)', color='blue', alpha=0.2,
                           ax=ax, data=df, order=['Boiling-2min', 'Boiling-5min', 'Boiling-15min', 'Pressure-3min'], orient='v')
        ax.set_xlabel("Staining Protocol")
      
        ax.xaxis.grid(True, linestyle='--', alpha=0.5)
        ax.yaxis.grid(True, clip_on=False)
    else:
        ax = sns.violinplot(y='Brown Intensity(25% top - 25% bottom)', x='Staining Protocol', inner='quartile',
                            linewidth=0.2, width=width, ax=ax, data=df,
                            order=['Boiling-2min', 'Boiling-5min', 'Boiling-15min', 'Pressure-3min'], orient='v')
        ax = sns.swarmplot(y='Brown Intensity(25% top - 25% bottom)', x='Staining Protocol', color='blue', alpha=0.2,
                           ax=ax, data=df, order=['Boiling-2min', 'Boiling-5min', 'Boiling-15min', 'Pressure-3min'], orient='v')
        ax.set_ylabel("Brown Intensity(25% top - 25% bottom)")
     
        ax.yaxis.grid(True, linestyle='--', alpha=0.5)
        ax.xaxis.grid(True, clip_on=False)
    
    return ax

# Assuming df is your DataFrame
ax = printViolin(df)
plt.savefig("brown_updated.pdf",format='pdf', dpi=1200)
plt.show()
```
